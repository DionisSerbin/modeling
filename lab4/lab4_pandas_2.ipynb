```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from matplotlib import pyplot as plt
import scipy.stats as stats
from sklearn.metrics import mean_squared_error
```

```python
def normalize(df):
    df_z_scaled = df.copy()

    for column in df_z_scaled.columns:
        if str(column) != 'id' and str(column) != 'product':
            df_z_scaled[column] = (df_z_scaled[column] -
                                   df_z_scaled[column].mean()) / df_z_scaled[column].std()
    return df_z_scaled
```

```python
df = normalize(df)
```

```python
df.hist(column='f0', bins=51)
```

```python
df_f0_clean = np.sqrt(df['f0'])

df_f0_clean.hist(bins=40)

mean = df_f0_clean.mean()
std = np.std(df_f0_clean)

plt.axvline(mean, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean - 3 * std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean + 3 * std, color='r', linestyle='dashed', linewidth=2)
```

```python
df.hist(column='f1', bins=40)

mean = df['f1'].mean()
std = np.std(df['f1'])

plt.axvline(mean, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean - 3 * std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean + 3 * std, color='r', linestyle='dashed', linewidth=2)
print(mean + 3 * std)
```

```python
df.hist(column='f2', bins=40)

mean = df['f2'].mean()
std = np.std(df['f2'])

plt.axvline(mean, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean - 3 * std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean + 3 * std, color='r', linestyle='dashed', linewidth=2)
print(mean + 3 * std)
```

```python
q1 = df["f0"].quantile(0.25)
q3 = df["f0"].quantile(0.75)
iqr = q3 - q1
df = df[~((df["f0"] < (q1 - 1.5 * iqr)) | (df["f0"] > (q3 + 1.5 * iqr)))]
df
```

```python
q1 = df["f1"].quantile(0.25)
q3 = df["f1"].quantile(0.75)
iqr = q3 - q1
df = df[~((df["f1"] < (q1 - 1.5 * iqr)) | (df["f1"] > (q3 + 1.5 * iqr)))]
df
```

```python
q1 = df["f2"].quantile(0.25)
q3 = df["f2"].quantile(0.75)
iqr = q3 - q1
df = df[~((df["f2"] < (q1 - 1.5 * iqr)) | (df["f1"] > (q3 + 1.5 * iqr)))]
df
```

```python
data_columns = ['f0', 'f1', 'f2']

df_target = df['product']
df_data = df[data_columns]

x_train, x_test, y_train, y_test = train_test_split(df_data, df_target, test_size=0.25, random_state=125)
x_train
```

```python
lin_regr = linear_model.LinearRegression()

lin_regr.fit(x_train, y_train)
predict = lin_regr.predict(x_test)
y_score = lin_regr.score(x_test, y_test)
mse = mean_squared_error(y_test, predict) ** 0.5
```

```python
print("Средний запас предсказанного на валидационной выборке сырья:", predict.mean(), '(тыс. баррелей)')
print(f"score: {y_score}")
print("RMSE модели линейной регрессии на валидационной выборке", mse)
```
