```python
def normalize(df):
    df_z_scaled = df.copy()

    for column in df_z_scaled.columns:
        if str(column) != 'id' and str(column) != 'product':
            df_z_scaled[column] = (df_z_scaled[column] -
                                   df_z_scaled[column].mean()) / df_z_scaled[column].std()
    return df_z_scaled
```

```python
REGION_BUDGET = 10_000_000_000
SEARCHING_WELLS = 500
BEST_WELLS = 200
PROFIT_PER_BARRELL = 450
PROFIT_PER_BARRELL_BY_PRODUCT = 450_000
PROB_DAMAGES = 0.025
VOLUME_FOR_PROFIT_WELL = REGION_BUDGET / BEST_WELLS / PROFIT_PER_BARRELL_BY_PRODUCT
VOLUME_FOR_PROFIT_WELL
```

```python
def get_best_wells(predict_product):
    return sorted(predict_product, reverse=True)[:BEST_WELLS]
```

```python
get_best_wells(predict)
```

```python
def get_profit(predict_product):
    profit = 0
    for product in predict_product:
        profit += product * PROFIT_PER_BARRELL_BY_PRODUCT
    return profit
```

```python
def avg(predict_product):
    return sum(predict_product) / len(predict_product)
```

```python
def check_damages(predict_product):
    profit = get_profit(predict_product)

    prob = profit / REGION_BUDGET

    if profit < REGION_BUDGET:
        return "  "

    if prob - 1 > PROB_DAMAGES:
        return "  "
    else:
        return "  "
```

```python
def get_average_profit(predict_product):
    return avg(predict_product) * PROFIT_PER_BARRELL_BY_PRODUCT
```

```python
check_damages(predict)
```
