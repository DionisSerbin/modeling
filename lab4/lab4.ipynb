```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None, 'display.width', -1)
df = pd.read_csv("place1.csv")
df
```

```python
df = df.drop(columns=['Unnamed: 5'])
df
```

```python
import scipy.stats as stats
q1 = df["f0"].quantile(0.25)
q3 = df["f0"].quantile(0.75)
iqr = q3 - q1

df = df[~((df["f0"] < (q1 - 1.5*iqr)) | (df["f0"] > (q3 + 1.5*iqr)))]
df
```

```python
q1 = df["f1"].quantile(0.25)
q3 = df["f1"].quantile(0.75)
iqr = q3 - q1

df = df[~((df["f1"] < (q1 - 1.5*iqr)) | (df["f1"] > (q3 + 1.5*iqr)))]
df
```

```python
q1 = df["f2"].quantile(0.25)
q3 = df["f2"].quantile(0.75)
iqr = q3 - q1

df = df[~((df["f2"] < (q1 - 1.5*iqr)) | (df["f2"] > (q3 + 1.5*iqr)))]
df
```

```python
from matplotlib import pyplot as plt

sqrt_df = (df)

sqrt_df.hist(column='f0', bins=39)

mean_value = sqrt_df['f0'].mean()
std = np.std(sqrt_df['f0'])

plt.axvline(mean_value, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean_value - 3 * std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean_value + 3 * std, color='r', linestyle='dashed', linewidth=2)
print(mean_value + 3 * std)
```

```python
from scipy.stats import pearsonr
```

```python
xs = df['f0']
ys = df['product'].apply(np.log)
pd.DataFrame(np.array([xs, ys]).T).plot.scatter(0, 1, s=12, grid=True)
plt.xlabel('f0')
plt.ylabel(' ')
plt.show()
print(df['f0'].corr(df['product']))
print(f" : {df['f0'].corr(df['product'])}")
print(f"  : {pearsonr(df['f0'], df['product'])}")
```
